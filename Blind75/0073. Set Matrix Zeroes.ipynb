## 0073. Set Matrix Zeroes

### method 1: A straightforward solution using O(mn) space is probably a bad idea.

```python
class Solution:
    def setZeroes(self, matrix):
        """
        Do not return anything, modify matrix in-place instead.
        """
        s = set() # use to store max of m * n coordinates
        ROWS = len(matrix)
        COLS = len(matrix[0])
        for i in range(ROWS):
            for j in range(COLS):
                if matrix[i][j] == 0:
                    s.add((i, j))
        while s: # set possible 0s
            x, y = s.pop()
            for i in range(ROWS):
                matrix[i][y] = 0
            for j in range(COLS):
                matrix[x][j] = 0
        return matrix
```

### A simple improvement uses O(m + n) space, but still not the best solution.

```python
class Solution:
    def setZeroes(self, matrix):
        """
        Do not return anything, modify matrix in-place instead.
        """
        ROWS = len(matrix)
        COLS = len(matrix[0])
        # use 2 sets to store x, y axis
        rows, cols = set(), set()
        for i in range(ROWS):
            for j in range(COLS):
                if matrix[i][j] == 0:
                    rows.add(i)
                    cols.add(j)
        # transform the set values to matrix
        for i in range(ROWS):
            for j in range(COLS):
                if i in rows or j in cols:
                    matrix[i][j] = 0
            
        return matrix
```

- use 2 sets stores values of x and y individually

### Could you devise a constant space solution?

```python
class Solution:
    def setZeroes(self, matrix):
        """
        Do not return anything, modify matrix in-place instead.
        """
        ROWS = len(matrix)
        COLS = len(matrix[0])
        # set 0 to inf is to avoid same 0s, coloring method
        for i in range(ROWS):
            for j in range(COLS):
                if matrix[i][j] == 0:
                    for ii in range(ROWS):
                        if matrix[ii][j] != 0: # shoold not set the unused 0s
                            matrix[ii][j] = float('inf') # set cols to inf
                    for jj in range(COLS):
                        if matrix[i][jj] != 0: # shoold not set the unused 0s
                            matrix[i][jj] = float('inf') # set row to inf
        # set inf to 0
        for i in range(ROWS):
            for j in range(COLS):
                if matrix[i][j] == float('inf'):
                    matrix[i][j] = 0
            
        return matrix
```
