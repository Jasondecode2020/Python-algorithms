## 133. Clone Graph

### method: dfs and adjacency list

```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []
"""

class Solution:
    def cloneGraph(self, node: 'Node') -> 'Node':
        clone = {}
        def dfs(node):
            if node in clone:
                return clone[node]
            copy = Node(node.val)
            clone[node] = copy # first clone the code
            for nei in node.neighbors: # then clone its neighbors
                copy.neighbors.append(dfs(nei))
            return copy # return the first cloned node addresss
                
        return dfs(node) if node else None
            
```
