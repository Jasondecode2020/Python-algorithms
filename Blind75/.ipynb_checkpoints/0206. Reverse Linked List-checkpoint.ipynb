## 0206. Reverse Linked List

### method 1: linked list reverse iteratively

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseList(self, head):
        prev = None
        while head:
            temp = head.next
            head.next = prev
            prev = head
            head = temp
        return prev
```

```python
s = Solution()
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
reverse = s.reverseList(head)
print('reversed linked list', reverse.val, reverse.next.val, reverse.next.next.val)
```

### method 2: linked list reverse recursively

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseList(self, head):
        def helper(prev, head):
            if not head: return prev
            temp = head.next
            head.next = prev
            prev = head
            head = temp
            return helper(prev, head)
        return helper(None, head)     
```

```python
s = Solution()
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
reverse = s.reverseList(head)
print('reversed linked list', reverse.val, reverse.next.val, reverse.next.next.val)
```

- for recursive, I prefer to use a helper funtion
