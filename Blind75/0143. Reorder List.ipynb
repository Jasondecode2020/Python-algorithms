## 0143. Reorder List

### method 1: linked list reverse iteratively

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reorderList(self, head):
        """
        Do not return anything, modify head in-place instead.
        """
        q = deque() # deque is useful api used in python
        node = head
        while node.next: # the q include without head
            node = node.next
            q.append(node)
        p = head
        while q:
            p.next = q.pop()
            p = p.next
            if q: # need to check if q exist after q.pop()
                p.next = q.popleft()
                p = p.next
        p.next = None
```

- there are many solutions, using a deque api maybe the easiest
