## 147. Insertion Sort List

### method 1

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def insertionSortList(self, head):
        '''
        Runtime: 4684 ms, faster than 5.03% of Python3 online submissions for Insertion Sort List.
        T = O(n^2), can be optimized using binary search n*log(n)
        '''
        dummy = ListNode(0.1)
        dummy.next = head
        temp = head
        while temp.next:
            temp = temp.next
            head = dummy.next
            while head:
                if temp == head:
                    break
                if head.val > temp.val:
                    head.val, temp.val = temp.val, head.val
                head = head.next
            
        return dummy.next
```

```python
s = Solution()
head = ListNode(4)
head.next = ListNode(2)
head.next.next = ListNode(1)
head.next.next.next = ListNode(3)
res = s.insertionSortList(head)
print(res.val)
print(res.next.val)
print(res.next.next.val)
print(res.next.next.next.val)
```
