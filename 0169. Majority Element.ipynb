## 169. Majority Element

### 1.1 method 1

```python
from collections import Counter
class Solution:
    def majorityElement(self, nums):
        # d = Counter(nums) # build a dict
        # for key in d:
        #     if d[key] > len(nums) // 2: # from keys find values
        #         return key
        n = len(nums)
        c = Counter(nums)
        for key in c:
            if c[key] > n // 2:
                return key
```

```python
s = Solution()
nums = [3,2,3]
s.majorityElement(nums)
```

### 1.2 method 2

#### Follow-up: Could you solve the problem in linear time and in O(1) space?
- 1 Prepare a dictionary, use only one loop
- 2 Use enumerate of index and values of nums array
- 3 Check is the numbers in the dict or not

```python
class Solution:
    def majorityElement(self, nums):
        count = 0
        majority_candidate = None
        for num in nums:
            if count == 0:
                majority_candidate = num
            if num == majority_candidate:
                count += 1
            else:
                count -= 1
        return majority_candidate
```

```python
s = Solution()
nums = [3,2,3]
s.majorityElement(nums)
```
