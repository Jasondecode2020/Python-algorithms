## 128. Longest Consecutive Sequence

### method 1: change structure
#### You must write an algorithm that runs in O(n^3) time.

```python
class Solution:
    def longestConsecutive(self, nums):
        maxN = 0
        for num in nums:
            nextN = num + 1
            currMaxN = 1
            while nextN in nums:
                currMaxN += 1
                nextN += 1
            maxN = max(currMaxN, maxN)
        return maxN
```

```python
s = Solution()
nums = [100,4,200,1,3,2]
s.longestConsecutive(nums)
```

### method 2: set
#### You must write an algorithm that runs in O(n^2) time.

```python
class Solution:
    def longestConsecutive(self, nums):
        numsset = set(nums)
        maxN = 0
        for num in nums:
            nextN = num + 1
            currMaxN = 1
            while nextN in numsset:
                currMaxN += 1
                nextN += 1
            maxN = max(currMaxN, maxN)
        return maxN
```

```python
s = Solution()
nums = [100,4,200,1,3,2]
s.longestConsecutive(nums)
```

### method 3: set
#### You must write an algorithm that runs in O(n) time.

```python
class Solution:
    def longestConsecutive(self, nums):
        numsset = set(nums)
        maxN = 0
        for num in nums:
            if num - 1 not in numsset:
                nextN = num + 1
                currMaxN = 1
                while nextN in numsset:
                    currMaxN += 1
                    nextN += 1
                maxN = max(currMaxN, maxN)
        return maxN
```

```python
s = Solution()
nums = [100,4,200,1,3,2]
s.longestConsecutive(nums)
```
