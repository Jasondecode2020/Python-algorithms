## 1812. Determine Color of a Chessboard Square

### method 1

```python
class Solution:
    def squareIsWhite(self, coordinates):
        s1 = coordinates[0]
        s2 = coordinates[1]
        if (ord(s1) - ord('a') + ord(s2) - ord('1')) % 2 == 0:
            return False
        return True
```

```python
s = Solution()
coordinates = "a1"
s.squareIsWhite(coordinates)
```
